```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Cost function for logistic regression
def compute_cost(X, y, theta):
    m = len(y)
    h = sigmoid(X @ theta)
    cost = (-1 / m) * (y.T @ np.log(h) + (1 - y).T @ np.log(1 - h))
    return cost[0, 0]  # Return as a scalar

# Gradient descent function for logistic regression
def gradient_descent(X, y, theta, learning_rate, num_iterations):
    m = len(y)
    cost_history = []

    for _ in range(num_iterations):
        h = sigmoid(X @ theta)
        gradient = (1 / m) * (X.T @ (h - y))  # Compute gradient
        theta -= learning_rate * gradient  # Update parameters
        cost = compute_cost(X, y, theta)  # Compute the cost
        cost_history.append(cost)

    return theta, cost_history

# Load dataset from CSV files
def load_dataset():
    # Load features (X) and target labels (y) from CSV files
    X = pd.read_csv('logisticX.csv', header=None).values
    y = pd.read_csv('logisticY.csv', header=None).values

    # Add bias term (column of 1s)
    X = np.hstack((np.ones((X.shape[0], 1)), X))  # Add bias term as the first column

    return X, y

# Visualize the dataset
def plot_dataset(X, y):
    plt.scatter(X[y.flatten() == 0][:, 1], X[y.flatten() == 0][:, 2], color="red", label="Class 0")
    plt.scatter(X[y.flatten() == 1][:, 1], X[y.flatten() == 1][:, 2], color="blue", label="Class 1")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend()
    plt.title("Dataset")
    plt.show()

# Main function
if __name__ == "__main__":
    # Load and visualize the dataset
    X, y = load_dataset()
    plot_dataset(X, y)

    # Initialize parameters
    theta = np.zeros((X.shape[1], 1))
    learning_rate = 0.01
    num_iterations = 1000

    # Perform gradient descent
    theta, cost_history = gradient_descent(X, y, theta, learning_rate, num_iterations)

    # Plot cost history
    plt.plot(range(len(cost_history)), cost_history, label="Cost")
    plt.xlabel("Iterations")
    plt.ylabel("Cost")
    plt.title("Cost Function Convergence")
    plt.legend()
    plt.show()

    # Final results
    print("Optimized theta:", theta)
    print("Final cost:", cost_history[-1])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, ConfusionMatrixDisplay

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Logistic regression model functions (same as earlier)
def compute_cost(X, y, theta):
    m = len(y)
    h = sigmoid(X @ theta)
    cost = -(1 / m) * (y.T @ np.log(h) + (1 - y).T @ np.log(1 - h))
    return cost

def gradient_descent(X, y, theta, learning_rate, num_iterations):
    m = len(y)
    cost_history = []

    for _ in range(num_iterations):
        h = sigmoid(X @ theta)
        gradient = (1 / m) * (X.T @ (h - y))
        theta -= learning_rate * gradient
        cost = compute_cost(X, y, theta)
        cost_history.append(cost)

    return theta, cost_history

# Load dataset (same as earlier)
def load_dataset():
    X = pd.read_csv('logisticX.csv', header=None).values
    y = pd.read_csv('logisticY.csv', header=None).values
    X = np.hstack((np.ones((X.shape[0], 1)), X))  # Add bias term as the first column
    return X, y

# Predict function
def predict(X, theta):
    return sigmoid(X @ theta) >= 0.5

# Main function
if __name__ == "__main__":
    # Load the dataset
    X, y = load_dataset()

    # Introduce two new independent variables (squares of the original features)
    X_new = np.hstack((X, X[:, 1:]**2))  # Add squared features to X

    # Initialize parameters
    theta = np.zeros((X_new.shape[1], 1))
    learning_rate = 0.01
    num_iterations = 1000

    # Train the logistic regression model
    theta, cost_history = gradient_descent(X_new, y, theta, learning_rate, num_iterations)

    # Make predictions on the dataset
    y_pred = predict(X_new, theta)

    # Compute confusion matrix
    cm = confusion_matrix(y, y_pred)
    print("Confusion Matrix:")
    print(cm)

    # Calculate accuracy, precision, recall, and F1-score
    accuracy = accuracy_score(y, y_pred)
    precision = precision_score(y, y_pred)
    recall = recall_score(y, y_pred)
    f1 = f1_score(y, y_pred)

    # Print the evaluation metrics
    print(f"\nAccuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")

    # Plot confusion matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
    disp.plot(cmap=plt.cm.Blues)
    plt.title("Confusion Matrix")
    plt.show()
```
